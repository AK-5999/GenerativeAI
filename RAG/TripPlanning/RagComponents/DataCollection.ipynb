```python
df.fillna("temp", inplace=True)
```

```python
df.shape
```

```python
output_list = []

destination_Tag = ""
for _, row in df.iterrows():
    destination = row['input__destination']
    interests = [
        row['input__interests__001'],
        row['input__interests__002'],
        row['input__interests__003'],
        row['input__interests__004'],
        row['input__interests__005'],
    ]
    state = row['output__optimized_itinerary__state']
    duration = row['input__duration']
    day = row['output__optimized_itinerary__days__day']
    morning = row['output__optimized_itinerary__days__morning']
    afternoon = row['output__optimized_itinerary__days__afternoon']
    evening = row['output__optimized_itinerary__days__evening']

    interests = [i for i in interests if i!="temp"]
    data = {
        destination: {
            "BestFor": interests,
            "State": state.replace("temp","N/A"),
            "Duration": duration,
            "Iternary": {
                day: {
                    "Morning": morning,
                    "Afternoon": afternoon,
                    "Evening": evening
                }
            }
        }
    }
    if destination != "temp":
        destination_Tag = destination
        output_list.append(data)
    else:
        orignal_data = output_list[-1]
        orignal_data[destination_Tag]["Iternary"][day] = {
                    "Morning": morning,
                    "Afternoon": afternoon,
                    "Evening": evening
                }
        output_list[-1] = orignal_data
```

```python
output_list
```

```python
# Convert list of dicts to one merged dict
merged_dict = {}
for item in output_list:
    merged_dict.update(item)

# Convert to JSON string
import json
json_output = json.dumps(merged_dict, indent=2)
```

```python
with open('output.json', 'w') as f:
    json.dump(merged_dict, f, indent=2)
```
